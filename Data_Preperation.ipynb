```
def prepare_image(file):
    img = cv2.imread(file)
    img = cv2.resize(img, (img_size, img_size))
    return img
```

```
from PIL import Image
import PIL.ImageOps
```

```
path1 = ("drive/My Drive/Hats_data/Train_Canny")
path3 = ("drive/My Drive/Hats_data/Train_Sketch")
import cv2
import os
from keras.preprocessing.image import array_to_img
import numpy as np
from matplotlib import pyplot as plt
img_size = 256 
for img in os.listdir(path):
  path_list = os.path.join(path,img)
  _img_ = prepare_image(path_list)
  _img_ = cv2.cvtColor(_img_, cv2.COLOR_BGR2RGB)
  canny = cv2.Canny(_img_, 300, 300)
  eds = canny.reshape(256,256,1)
  sob = array_to_img(eds)
  path2 = os.path.join(path1, img)
  sob.save(path2)
  inverted_image = PIL.ImageOps.invert(sob)
  path4 = os.path.join(path3, img)
  inverted_image.save(path4)
```

```
path6 = ("drive/My Drive/Hats_data/Train_Images_256")
for img in os.listdir(path):
  path_list = os.path.join(path,img)
  _img_ = prepare_image(path_list)
  _img_ = array_to_img(_img_)
  _img_.save(os.path.join(path6,img))
```
